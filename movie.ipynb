```python
import numpy as np
import pandas as pd
import ast

movies=pd.read_csv('tmdb_5000_movies.csv')
credits=pd.read_csv('tmdb_5000_credits.csv')

movies=movies.merge(credits,on='title')

movies.head()

movies=movies[['movie_id','title','overview','genres','keywords','cast','crew']]

movies.isnull().sum()

movies.dropna(inplace=True)

def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L

movies['genres'].apply(convert)

movies['genres']=movies['genres'].apply(convert)

movies['keywords']=movies['keywords'].apply(convert)

movies.head()

def convertcast(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if(counter!=3):
            L.append(i['name'])
            counter+=1;
        else:
            break
    return L

movies['cast']=movies['cast'].apply(convertcast)

movies['cast']

movies['crew']

def convertd(obj):
    L=[]
    for i in ast.literal_eval((obj)):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L

movies['crew'].apply(convertd)

movies['crew']=movies['crew'].apply(convertd)

movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","")for i in x])

movies.head()

movies['overview']=movies['overview'].apply(lambda x:x.split())

movies.head()

movies['overview']=movies['overview'].apply(lambda x:[i.replace(" ","")for i in x])
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","")for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","")for i in x])

movies.head()

movies.head()

movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']

movies['tags']

new_df=movies[['movie_id','title','tags']]

new_df.head()

new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))

new_df.head()
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
import nltk
from sklearn.feature_extraction.text import CountVectorizer
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
cv=CountVectorizer(max_features=5000,stop_words='english')
def stem(text):
    L=[]
    for i in text.split():
        L.append(ps.stem(i))
    return " ".join(L)
new_df['tags']=new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similar=cosine_similarity(vectors)
```

```python
similar.shape
```

```python
similar.shape
```

```python
def recommend(movie):
    movie_i=new_df[new_df['title']==movie].index[0]
    distances=similar[movie_i]
    movies_l=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movies_l:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
